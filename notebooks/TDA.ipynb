Преобразуем считанные данные из 4 и 5 столбцов в соответствующие списки.

```python
def returnRightLists(my_df):
    for i, k in enumerate(my_df['SentenceList']):
      sent_list = re.split(r',', k[1:-1])
      sent_list = [s.strip()[1:-1] for s in sent_list]
      my_df['SentenceList'][i] = sent_list

    for i, k in enumerate(my_df['TokensList']):
      tokens_list = list()
      my_list = re.split(r'],', k[1:-2])
      my_list = [s.strip()[1::] for s in my_list]
      for m in my_list:
        tokens = re.split(r',', m)
        tokens = [t.strip()[1:-1] for t in tokens]
        tokens_list.append(tokens)
      my_df['TokensList'][i] = tokens_list

    return my_df
```

Удалю предложения, в которых всего 4 слова (часто было предложение - федеральных законов).

```python
import copy

def cleaning_sentence(dataframe):
  for i in range(len(dataframe)):
    counter = 0
    tokens = copy.deepcopy(dataframe['TokensList'][i])
    sentences = copy.deepcopy(dataframe['SentenceList'][i])

    for i_t in range(len(dataframe['TokensList'][i])):
      if len(dataframe['TokensList'][i])>1:
        if len(dataframe['TokensList'][i][i_t])<5:
          tokens.pop(i_t-counter)
          sentences.pop(i_t-counter)
          counter+=1

    dataframe['TokensList'][i]=tokens
    dataframe['SentenceList'][i]=sentences

  return dataframe
```

### Reading files

```python
def get_pdf_text(file):
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    converter = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(file, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, converter)
    password = ""
    maxpages = 0
    caching = True
    pagenos = set()
    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password, caching=caching, check_extractable=True):
        interpreter.process_page(page)
    fp.close()
    converter.close()
    str = retstr.getvalue()
    retstr.close()
    return str

def get_docx_text(file):
  doc = docx.Document(file)
  text = []
  for paragraph in doc.paragraphs:
    text.append(" " + paragraph.text)
  return text

def preprocess_text(text):
  'Дописать функцию!!!'
  #def remove_point(text):
    #sub_str = ["в ред\.", ]
    #return
  punctuation = '"#$&*+/-<=>@[\\]^_`{|}~•'
  # remove punctuation marks
  text = ''.join(ch for ch in text if ch not in set(punctuation))
  # replace new lines
  text = text.replace('\n', ' ') 
  text = text.strip()
  # remove dates
  text = text.replace(r'(\d+\.\d+\.\d+)', "")
  # remove numbers
  text = re.sub(r'\d', '', text)
  # remove '.' from 'в ред.'
  text = text.replace("в ред\.", "в ред")
  # remove '.' from 'см.'
  text = text.replace("см\.", "см")
  # remove '.' from 'ст.'
  text = text.replace("ст\.", "ст")
  # remove '.' from 'гл.'
  text = text.replace("гл\.", "гл")
  # remove '.' from 'разд.'
  text = text.replace("разд\.", "разд")
  # remove '.' from 'т.е.'
  text = text.replace("т\.е\.", "те")
  # remove '.' from 'т.к.'
  text = text.replace("т\.к\.", "тк")
  # remove '.' from 'т.ч.'
  text = text.replace("т\.ч\.", "тч")
  # remove '.' from 'т.д.'
  text = text.replace("т\.д\.", "тд.")
  # remove '.' from 'ч.'
  text = text.replace("ч\.", "ч")
  # remove '.' from 'п.'
  text = text.replace("п\.", "п")
  # remove '.' from 'утв.'
  text = text.replace("утв\.", "утв")
  # remove '.' from 'коммент.'
  text = text.replace("коммент\.", "коммент")
  # convert text to lowercase
  text = text.lower()
  # add spaces before punctuation symbols
  text = re.sub('([.,;:!?()])', r' \1 ', text)
  # remove duplicating whitespaces
  text = re.sub(' +', ' ', text)

  return text
```

## *Kepler Mapper*

```python
article_exception = []
```

```python
def KeplerMapperVisualise(ind, ind2 = 0, filename='', title='', show=False):
    '''Returns the simplicial_complex - a dictionary with “nodes”, “links” and “meta” information'''
    if title=='':
      title = "Статья №" + str(df['Article №'][ind])
    if df['NumSentences'][ind]<3:
      return 0
    # Initialize
    mapper = km.KeplerMapper(verbose=0)
    # Fit to and transform the data
    if ind2 ==0:
      data_array = df['Embeddings'][ind]
      sentences = np.array([str(i) + '. ' + s for i, s in enumerate(df['SentenceList'][ind])])
      n=3
      o = 0.65
      if df['NumSentences'][ind]<6:
        #n = 2
        num_clusters = 2
      elif 5<df['NumSentences'][ind]<15:
        #n=5
        num_clusters = 3
      elif df['NumSentences'][ind]>14:
        #n=10
        num_clusters = 4
    else:
      data_array = []
      for emb in df['Embeddings'][ind:ind2]:
        for e in emb:
          data_array.append(e)
      n = 35
      o = 0.7
    try:
      projected_data = mapper.fit_transform(data_array,projection = TSNE(n_components=2, metric = 'cosine'))
      print("Shape of projected data: ", projected_data.shape)
      # Create dictionary called 'graph' with nodes, edges
      graph = mapper.map(projected_data, clusterer=cluster.AgglomerativeClustering(n_clusters=num_clusters,
                                                              linkage="complete" , # complete, average
                                                              affinity="cosine"),
                            cover=km.Cover(n_cubes=n, #35, 10, 25, 3
                                          perc_overlap=o), #0.9, 0.7, 0.9, 0.7/0.9
                                          remove_duplicate_nodes=True)
      # Visualize it
      if filename=='':
        path = 'drive/My Drive/Thesis/KeplerMapperPictures/'+'Статья_'+ str(df['Article №'][ind]) +'.html'
      else:
        path = 'drive/My Drive/Thesis/KeplerMapperPictures/'+ filename + '.html'
      html = mapper.visualize(graph, path_html=path, title=title, custom_tooltips=sentences)
    except Exception:
      article_exception.append(ind)
    if show:
        km.drawing.draw_matplotlib(graph)
```

## *Persistence diagrams*

```python
def PersDiagram(ind, title=''):
  reducer = umap.UMAP(metric='cosine')
  projected_data = reducer.fit_transform(df['Embeddings'][ind])
  if title=='':
    title = "Статья №" + str(df['Article №'][ind])
  diagram = ripser(projected_data)['dgms']
  plot_diagrams(diagram, title= title, show=True)
```

```python
def BottleneckMatching(ind1, ind2):
  reducer = umap.UMAP(metric='cosine')
  projected_data1 = reducer.fit_transform(df['Embeddings'][ind1])
  title1 = "Статья №" + str(df['Article №'][ind1])
  diagram1 = ripser(projected_data1)['dgms'][1]

  projected_data2 = reducer.fit_transform(df['Embeddings'][ind2])
  title2 = "Статья №" + str(df['Article №'][ind2])
  diagram2 = ripser(projected_data2)['dgms'][1]

  d, (matching, D) = bottleneck(diagram1, diagram2, matching=True)
  bottleneck_matching(diagram1, diagram2, matching, D, labels=[title1, title2])
  plt.title("Расстояние {:.3f}".format(d))
  plt.show()
```
